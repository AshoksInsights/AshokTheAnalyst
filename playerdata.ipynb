```python
player_detail=[]
for i in Player:
  try:
      url="https://www.espncricinfo.com/cricketers/"+str(i)
      response=get(url,headers=headers)
      soup=BeautifulSoup(response.text,'html.parser')
      header=soup.find_all('p',class_="ds-text-tight-m ds-font-regular ds-uppercase ds-text-typo-mid3")
      value=soup.find_all('span',class_="ds-text-title-s ds-font-bold ds-text-typo")
      header_=[]
      value_=[]
      for x in range(len(header)):
             a=header[x].text
             b=value[x].text
             header_.append(a)
             value_.append(b)
      df=pd.DataFrame([value_],columns=header_)
      df['id'] = i.split("-")[-1]  
      player_detail.append(df)
  except IndexError:
      break
```

```python
Player_detail = pd.concat(player_detail)
```

```python
Player_detail.to_csv("Player_detail.csv")
```
